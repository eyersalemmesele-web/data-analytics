Monthly Revenue Analysis

What is the monthly total revenue for in-store sales in each of the two sales territories over the full period covered by the data?

I compared monthly and yearly sales for two managers to see how revenue changes over time.

```python
# Convert date column
store_sales["Transaction Date"] = pd.to_datetime(store_sales["Transaction Date"])

# Create YearMonth
store_sales["YearMonth"] = store_sales["Transaction Date"].dt.to_period("M")

# Merge datasets
sales_detail = store_sales.merge(store_detail, on="Store ID", how="left")

# Filter territories
erbayne_data = sales_detail[sales_detail["Territory Manager"] == "Erbayne Middleton"]
shruti_data = sales_detail[sales_detail["Territory Manager"] == "Shruti Reddy"]

# Monthly revenue
erbayne_monthly = erbayne_data.groupby("YearMonth")["Sale Amount"].sum().reset_index()
shruti_monthly = shruti_data.groupby("YearMonth")["Sale Amount"].sum().reset_index()

# Rename
erbayne_monthly.columns = ["YearMonth", "Total Revenue"]
shruti_monthly.columns = ["YearMonth", "Total Revenue"]

erbayne_monthly
shruti_monthly
```

```python
# Load customer dataset (pipe-separated file)
import pandas as pd

customer_list = pd.read_csv("customer_list.csv", sep="|")

# Merge customer data with sales data using RewardsID
# This connects each purchase to a specific customer
customer_sales = pd.merge(
    customer_list,
    store_sales,
    left_on="cust_id",
    right_on="RewardsID",
    how="inner"
)

# Filter customers by Erbayne Middleton's territory stores
erbayne_customers = customer_sales[
    customer_sales["Store ID"].isin(erbayne_stores["Store ID"])
]

# Filter customers by Shruti Reddy's territory stores
shruti_customers = customer_sales[
    customer_sales["Store ID"].isin(shruti_stores["Store ID"])
]

# Calculate total spending per customer (Erbayne)
erbayne_top = erbayne_customers.groupby(
    ["cust_id", "name"]
)["Sale Amount"].sum().reset_index()

erbayne_top = erbayne_top.sort_values(
    "Sale Amount",
    ascending=False
).head(10)

# Calculate total spending per customer (Shruti)
shruti_top = shruti_customers.groupby(
    ["cust_id", "name"]
)["Sale Amount"].sum().reset_index()

shruti_top = shruti_top.sort_values(
    "Sale Amount",
    ascending=False
).head(10)

# Show results
print("Top Customers - Erbayne Middleton")
display(erbayne_top)

print("Top Customers - Shruti Reddy")
display(shruti_top)
```

# ANALYSIS QUESTION 5

Product Categories

What is the number of transactions per month by product category in each assigned territory? What is total sales revenue per month by category? What might this tell us about the most popular products, and where could there be opportunity for growth?

I joined sales, product, and category datasets to understand performance across different product groups.

```python
# Combine Erbayne Middleton sales with product and category information
erbayne_products = erbayne_sales.merge(products, on="Prod Num", how="inner")

erbayne_products = erbayne_products.merge(
    product_categories[["CategoryID", "Category"]],
    on="CategoryID",
    how="left"
)

# Combine Shruti Reddy sales with product and category information
shruti_products = shruti_sales.merge(products, on="Prod Num", how="inner")

shruti_products = shruti_products.merge(
    product_categories[["CategoryID", "Category"]],
    on="CategoryID",
    how="left"
)

# Preview results
display(erbayne_products.head())
display(shruti_products.head())
```

This step merges sales data with product and category information for both territories. This allows each transaction to be analyzed based on product type and category.

By combining these datasets, we can better understand which products and categories contribute most to sales in each territory.

```python
# Function to get category summary
def category_summary(data):
    summary = data.groupby("Category")["Sale Amount"].agg(
        Transactions="count",
        Total_Revenue="sum"
    ).reset_index()

    return summary.sort_values("Total_Revenue", ascending=False)

# Erbayne results
erbayne_cat = category_summary(erbayne_products)
display(erbayne_cat)

# Shruti results
shruti_cat = category_summary(shruti_products)
display(shruti_cat)
```

This analysis examines product categories by calculating both the number of transactions and total revenue for each category in both territories.

The goal is to understand which product categories are most popular (based on transactions) and which generate the most revenue.

The data was grouped by category to compare product performance across Erbayne Middleton and Shruti Reddy’s territories. This helps identify trends in customer purchasing behavior and potential opportunities for growth in underperforming categories.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

# Make sure months are sorted (VERY IMPORTANT)
category_summary = category_summary.sort_values("Month")

for category in category_summary["Category"].unique():
    data = category_summary[category_summary["Category"] == category]

    plt.plot(
        data["Month"].astype(str),
        data["Total Revenue"],
        marker="o",   # makes it easier to see points
        label=category
    )

plt.title("Monthly Revenue by Product Category")
plt.xlabel("Month")
plt.ylabel("Total Revenue ($)")
plt.xticks(rotation=45)

plt.legend(title="Category")
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

This visualization shows monthly revenue trends by product category. Each line represents a category and shows how revenue changes over time.

This helps identify which categories are consistently strong and which fluctuate, indicating possible seasonal demand or growth opportunities.

I used a line chart to compare monthly revenue trends across product categories.

This line chart visualizes monthly revenue trends by product category to analyze performance patterns and identify growth opportunities over time.

# Additional Analysis: Top Product Category Revenue Trends

In this section, I analyze the two highest-performing product categories: Technology & Accessories and Textbooks.

The goal is to understand:
- How revenue changes over time
- Which category performs better
- How each region contributes to performance

```python
# Additional Analysis: Top Product Category Revenue Trends

# Merge all sales with product and category information
all_products = store_sales.merge(products, on='Prod Num')
all_products = all_products.merge(
    product_categories[['CategoryID', 'Category']],
    on='CategoryID'
)

# Convert date and create quarter column
all_products['Transaction Date'] = pd.to_datetime(all_products['Transaction Date'])
all_products['Quarter'] = all_products['Transaction Date'].dt.to_period('Q')

# Focus on top 2 categories
top_categories = ['Technology & Accessories', 'Textbooks']

# Quarterly revenue for top categories (all regions)
all_quarterly_cat = all_products[
    all_products['Category'].isin(top_categories)
].groupby(['Quarter', 'Category'])['Sale Amount'].sum().reset_index()

# Rename column for clarity
all_quarterly_cat = all_quarterly_cat.rename(
    columns={'Sale Amount': 'Total Revenue'}
)

# Show result
print(all_quarterly_cat.to_string(index=False))
```

In this additional analysis, I combined sales, product, and category data to analyze quarterly revenue trends for the top two product categories: Technology & Accessories and Textbooks.

This helps identify how the most important categories perform over time and supports deeper understanding of overall revenue patterns across the business.

```python
# Quarterly Revenue for Top Categories (Erbayne vs Shruti)

top_cats = ['Technology & Accessories', 'Textbooks']

# Ensure Quarter column exists
erbayne_products['Quarter'] = pd.to_datetime(erbayne_products['Transaction Date']).dt.to_period('Q')
shruti_products['Quarter'] = pd.to_datetime(shruti_products['Transaction Date']).dt.to_period('Q')

# Erbayne quarterly revenue
erbayne_quarterly_top = erbayne_products[
    erbayne_products['Category'].isin(top_cats)
].groupby(['Quarter', 'Category'])['Sale Amount'].sum().reset_index()

erbayne_quarterly_top = erbayne_quarterly_top.rename(
    columns={'Sale Amount': 'Total Revenue'}
)

# Shruti quarterly revenue
shruti_quarterly_top = shruti_products[
    shruti_products['Category'].isin(top_cats)
].groupby(['Quarter', 'Category'])['Sale Amount'].sum().reset_index()

shruti_quarterly_top = shruti_quarterly_top.rename(
    columns={'Sale Amount': 'Total Revenue'}
)

# Show results
print("Erbayne Quarterly Revenue (Top Categories)")
print(erbayne_quarterly_top.to_string(index=False))

print("\nShruti Quarterly Revenue (Top Categories)")
print(shruti_quarterly_top.to_string(index=False))
```

```python
# Standard libraries 
import pandas as pd          # data manipulation — tables, merging, grouping
import numpy as np           # numbers and arrays
import matplotlib.pyplot as plt          # charts and graphs
import matplotlib.ticker as mticker      # custom axis labels (e.g. $1.2M)
import matplotlib.patches as mpatches   # custom legend items
import warnings
warnings.filterwarnings('ignore')        # keep output clean

from IPython.display import display, HTML   # show colored tables in Jupyter

# Chart colour palette 
# Every chart uses these same colours so the notebook looks consistent
EM_COLOR   = '#68d391'   # green  — Erbayne Middleton
SR_COLOR   = '#63b3ed'   # blue   — Shruti Reddy
DANGER     = '#fc8181'   # red    — revenue drops / low performers
ACCENT     = '#f6ad55'   # orange — #1 ranked item highlight
HIGH_GREEN = '#276749'   # dark green background — highest value in table
LOW_RED    = '#742a2a'   # dark red background   — lowest value in table
PURPLE     = '#b794f4'
TEAL       = '#76e4f7'
YELLOW     = '#faf089'
CAT_COLORS = [EM_COLOR, SR_COLOR, ACCENT, DANGER, PURPLE, TEAL, YELLOW]

# USD axis formatter 
# Used on chart axes so $1,200,000 becomes $1.2M
def fmt_usd(x, _=None):
    if abs(x) >= 1e6: return f'${x/1e6:.1f}M'
    if abs(x) >= 1e3: return f'${x/1e3:.0f}K'
    return f'${x:.0f}'

# Colored HTML table helper
# highlight_col   = column name whose numeric value drives row colour
# high_is_good    = True  → highest value = green,  lowest = red
#                   False → highest value = red,     lowest = green  (e.g. dip %)
def show_table(df, title='', max_rows=30, highlight_col=None, high_is_good=True):
    """Render a DataFrame as a styled HTML table.
    - Dark navy header row
    - Alternating dark rows
    - Optional: highest value row highlighted GREEN, lowest highlighted RED
      (or reversed when high_is_good=False, e.g. for % drops)
    """
    th_style = ('background:#1a365d;color:#bee3f8;padding:8px 14px;'
                'border:1px solid #2d3748;font-size:12px;text-align:left')
    td_a  = 'background:#1a202c;color:#e2e8f0;padding:7px 12px;border:1px solid #2d3748'
    td_b  = 'background:#171f2a;color:#e2e8f0;padding:7px 12px;border:1px solid #2d3748'
    td_hi = ('background:#1c4532;color:#9ae6b4;padding:7px 12px;'
             'border:1px solid #2d3748;font-weight:bold')   # green — best
    td_lo = ('background:#742a2a;color:#feb2b2;padding:7px 12px;'
             'border:1px solid #2d3748;font-weight:bold')   # red  — worst

    cap = (f"<p style='font-weight:bold;font-size:14px;color:#63b3ed;"
           f"margin-bottom:4px'>{title}</p>") if title else ''

    # Find which row index is highest / lowest (numeric col only)
    hi_idx = lo_idx = None
    if highlight_col and highlight_col in df.columns:
        nums = pd.to_numeric(df[highlight_col], errors='coerce')
        if nums.notna().any():
            hi_idx = nums.idxmax()
            lo_idx = nums.idxmin()
            if not high_is_good:
                hi_idx, lo_idx = lo_idx, hi_idx  # flip for bad metrics

    header = '<tr>' + ''.join(f'<th style="{th_style}">{c}</th>' for c in df.columns) + '</tr>'
    rows_html = ''
    for i, (idx, row) in enumerate(df.head(max_rows).iterrows()):
        if idx == hi_idx:
            row_td = td_hi
        elif idx == lo_idx:
            row_td = td_lo
        else:
            row_td = td_a if i % 2 == 0 else td_b
        rows_html += '<tr>' + ''.join(f'<td style="{row_td}">{v}</td>' for v in row) + '</tr>'

    html = (f"<div style='margin:12px 0'>{cap}"
            f"<table style='border-collapse:collapse;width:100%'>"
            f"<thead>{header}</thead><tbody>{rows_html}</tbody></table></div>")
    display(HTML(html))

# Dark theme for ALL charts
plt.rcParams.update({
    'figure.facecolor': '#0f1117',
    'axes.facecolor':   '#1a202c',
    'axes.edgecolor':   '#2d3748',
    'axes.labelcolor':  '#e2e8f0',
    'xtick.color':      '#a0aec0',
    'ytick.color':      '#a0aec0',
    'text.color':       '#e2e8f0',
    'grid.color':       '#2d3748',
    'legend.facecolor': '#1a202c',
    'legend.edgecolor': '#2d3748',
    'font.size':        10,
})

print(' Setup complete — libraries loaded, dark theme applied.')
```
