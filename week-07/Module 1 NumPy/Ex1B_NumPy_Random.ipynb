# Exercise 1.B NumPy Random

# Import NumPy

```python
import numpy as np

# Imported NumPy library as np for random number generation
```

# Generating Random Numbers
# Random integers (size of 2)

```python
r_1 = np.random.randint(1, 20, size=2)

print(r_1)

# Generated 2 random integers between 1 and 20
```

# Random integers (1 row, 2 columns)

```python
r_2 = np.random.randint(1, 20, size=(1, 2))

print(r_2)

# Generated a 1x2 array of random integers
```

# Random floats (2 rows, 1 column)

```python
r_3 = np.random.rand(2, 1)

print(r_3)

# Generated random float numbers between 0 and 1
```

# Random integers (4 rows, 5 columns)

```python
r_4 = np.random.randint(20, 100, size=(4, 5))

print(r_4)

# Generated a 4x5 array of random integers between 20 and 100
```

# Random floats (4 rows, 5 columns)

```python
r_5 = np.random.rand(4, 5)

print(r_5)

# Generated a 4x5 array of random float values
```

# Generating Random Choices from an Array
# Create students array

```python
students = np.array([
    "Eyerusalem", "Selam", "Mina", "Shewit",
    "Daniel", "Max", "Jerry", "Hiwot",
    "Nahom", "Almaz", "Noah", "Heaven"
])

print(students)

# Created a 1D array with 12 student names
```

# Randomly call on one student

```python
call_on = np.random.choice(students)

print(call_on)

# Randomly selected one student from the array
```

# Create breakout groups

```python
breakout_groups = np.random.choice(students, size=(3, 4))

print(breakout_groups)

# This may not be the best way to create groups because
# some students can appear more than once while others may not appear at all.
# Generated random breakout groups
```

# Weighted random selections

```python
call_on_weighted = np.random.choice(
    students,
    size=5,
    p=[0.30, 0.10, 0.05, 0.10,
       0.05, 0.10, 0.00, 0.05,
       0.05, 0.10, 0.05, 0.05]
)

print(call_on_weighted)

# Alex has a 30% chance of being selected.
# Michael has a 0% chance of being selected.
# Other students were given smaller probability values.
# Used weighted probabilities for random student selection
```

# Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))

print(bell_curve)

# Generated a normal distribution with mean 10 and standard deviation 5
```
