# Exercise 1.C Performing operations on arrays

# Import NumPy

```python
import numpy as np

# Imported NumPy as np for array operations
```

# Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
r_2 = np.random.randint(1, 20, size=(1, 2))
r_3 = np.random.rand(2, 1)
r_4 = np.random.randint(20, 100, size=(4, 5))
r_5 = np.random.rand(4, 5)

print(r_1)
print(r_2)
print(r_3)
print(r_4)
print(r_5)

# Created sample random arrays for use in operations
```

# Permutations

```python
print(r_4)
np.random.shuffle(r_4)
print(r_4)

# shuffle changes the array in place by rearranging rows randomly
```

```python
new_shuffle = np.random.permutation(r_4)

print(new_shuffle)
print(r_4)

# permutation returns a new shuffled array and does NOT change original
# Shuffle vs permutation:
# shuffle modifies original array, permutation creates a copy
```

# Mathematical Operations
# r_1 and r_2

```python
print(r_1)
print(r_2)

print(r_1.shape)
print(r_2.shape)

print(np.add(r_1, r_2))
print(np.add(r_1, r_2).shape)

# Arrays broadcast because shapes are compatible
```

# r_1 and r_3

```python
print(r_1)
print(r_3)

print(r_1.shape)
print(r_3.shape)

print(np.add(r_1, r_3))
print(np.add(r_1, r_3).shape)

# Broadcasting expands smaller array to match shape
```

# r_2 and r_4

```python
print(r_2)
print(r_4)

print(r_2.shape)
print(r_4.shape)

print(np.add(r_2, r_4))
print(np.add(r_2, r_4).shape)

# Broadcasting works because r_2 can expand to match r_4 rows/columns
```

# Invalid operations

```python
np.add(r_1, r_4)
# Error: shapes are not compatible for broadcasting

# r_1, r_3, and r_4 cannot all broadcast together due to shape mismatch
```

# Scalar / small array broadcasting

```python
print(np.add(np.array([1]), r_4))

# Scalar-like array broadcasts across all elements
```

```python
print(np.add(np.array([[1]]), r_4))

# 1x1 array broadcasts to full shape of r_4
```

```python
print(np.add(np.array([1,1,1,1,1]), r_4))

# 1D array matches columns and broadcasts across rows
```

```python
np.add(np.array([[1],[1]]), r_4)

# Error: shape mismatch, cannot broadcast 2x1 with 4x5 array
# Broadcasting explanation:
# NumPy expands smaller arrays to match larger shapes automatically
# The result shape is determined by the largest compatible shape
```

# Rounding

```python
print(r_3)

# Display original float values
```

```python
print(np.trunc(r_3))

# Removes decimal part (cuts off values)
```

```python
print(np.fix(r_3))

# Similar to trunc, removes decimals toward zero
```

```python
print(np.floor(r_3))

# Rounds down to nearest integer
```

```python
print(np.ceil(r_3))


# Rounds up to nearest integer
```

```python
print(np.round(r_3, 2))

# Rounds values to 2 decimal places
```

# Cumulative Operations

```python
print(r_4)

# Display r_4 for calculations
```

```python
print(np.sum(r_4))

# Total sum of all elements
```

```python
print(np.sum(r_4, axis=0))

# Sum of each column
```

```python
print(np.sum(r_4, axis=1))


# Sum of each row
```

```python
print(np.cumsum(r_4))

# Running total of all elements
```

```python
print(np.cumsum(r_4, axis=0))

# Running total by column
```

# Grocery Problem

```python
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)

# Each index represents: vegetables, fruits, snacks, beverages
```

```python
print("The first bag contains", np.sum(bag1), "items, and the second bag contains", np.sum(bag2), "items.")
```

```python
total_categories = bag1 + bag2

print("There are", total_categories[0], "vegetables,", 
      total_categories[1], "fruits,", 
      total_categories[2], "boxes of snacks, and", 
      total_categories[3], "beverages in the two grocery bags.")
```

# Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])
all_knocks = np.array([ali, ben, carla])

print(all_knocks)

# Door knock data for each team member (Mon–Fri)
```

```python
print("Ali total:", np.sum(ali))
print("Ben total:", np.sum(ben))
print("Carla total:", np.sum(carla))

print("Group total:", np.sum(all_knocks))

```

```python
daily_total = np.sum(all_knocks, axis=0)
estimated_votes = daily_total / 2

print(estimated_votes)

# Estimated votes based on rule: 2 door knocks = 1 vote
```
