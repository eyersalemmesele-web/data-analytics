# Exercise 1.A NumPy: Manipulating Arrays

# Importing NumPy

```python
import numpy as np

 # Imported NumPy library as np for easier array operations
```

# Looping Through Arrays
# Create arrA

```python
arrA = np.array([(11, 12, 13), (14, 15, 16)])
print(arrA)

# Created a 2D NumPy array with two rows of values
```

# Loop through first dimension

```python
for x in arrA:
    print(x * 2)
    
# Multiplies each row by 2
```

# Loop through second dimension

```python
for x in arrA:
    for y in x:
        print(y * 2)

# This loop goes element by element and multiplies each number individually.

#The first loop works on whole rows, so it prints arrays.
#The second loop works on single values, so it prints individual numbers.
```

# Loop result variable

```python
for x in arrA:
    loop_result = x * 2
    print(loop_result)

# x is a row, so multiplying it by 2 doubles all values in that row
# Each row is multiplied at once because NumPy applies operations to the whole array automatically.
```

# Joining and Stacking Arrays
# Concatenate axis=0

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)

# This adds the arrays vertically (adds more rows).
```

# Concatenate axis=1

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)

# This adds the arrays horizontally (adds more columns).
```

# Stack arrays

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)

# This creates a new dimension by stacking arrays together.
```

# Stack error example

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)


# Cannot concatenate because array dimensions do not match along axis 1 (different number of columns)
# This fails because the arrays do not have the same shape.
```

# Another concatenate

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)

# This combines arrA back into the larger array vertically.
```

# Axis=1 error

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)

# Error: arrays must have same number of rows
# This fails because the arrays do not match in shape along axis 1.
```

# hstack, vstack, dstack

```python
h = np.hstack((arrA, arrA))
v = np.vstack((arrA, arrA))
d = np.dstack((arrA, arrA))

print(h)
print(v)
print(d)

# hstack → joins columns
# vstack → joins rows
# dstack → adds a third dimension
```

# Splitting Arrays
# Split into 2

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)

#This splits the array into 2 parts.
```

# Split into 3

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)

#This splits the array into 3 parts.
```

# Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, "odd")
print(result_1)

#Even numbers stay the same, odd numbers are replaced with "odd".
```

# Sorting and Filtering Arrays
# Create arrB and sort

```python
arrB = np.array([[7, 6],
                 [13, 5],
                 [7, 3]])

arrB_sorted = np.sort(arrB)
print(arrB_sorted)

# Each row is sorted in ascending order.
```

# Filter values less than 7

```python
print(arrB)

z = arrB < 7
arrC = arrB[z]

print(arrC)

# This keeps only values less than 7.
```

# Filter values greater or equal to 7

```python
y = arrB >= 7
arrD = arrB[y]

print(arrD)

# This keeps only values greater than or equal to 7.
```
