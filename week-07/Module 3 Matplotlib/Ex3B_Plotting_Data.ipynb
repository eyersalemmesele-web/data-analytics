# Advanced Plotting Techniques

```python
import matplotlib.pyplot as plt
```

# Bar Chart

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales, color='skyblue', width=0.5)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.show()

# Created a bar chart to show sales for different products
# Customized color and bar width for better visualization
```

# Scatter Plot

# Subplots

# Line + Histogram Layout

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Created a figure with 2 subplots side by side
```

# Sine Wave (First Plot)

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax[0].plot(x, y)
ax[0].set_title("Sine Wave")
ax[0].set_xlabel("X values")
ax[0].set_ylabel("Y values")

# Plotted sine wave in first subplot
```

# Histogram (Second Plot)

```python
ax[1].hist(np.random.randn(1000))
ax[1].set_title("Histogram")
ax[1].set_xlabel("Values")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()

# Created histogram of 1000 random values in second subplot
# Used tight_layout to prevent overlap
```
