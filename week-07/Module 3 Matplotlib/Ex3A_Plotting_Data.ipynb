# Exercise 3.A Getting familiar with Matplotlib

# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Imported NumPy for data handling and Matplotlib for plotting
```

# Creating Basic Plots
# Create data

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Created x and y data points for plotting
```

# Basic line plot

```python
plt.plot(x, y)
plt.show()

# Created a basic line graph using x and y values
```

# Plot with markers

```python
plt.plot(x, y, marker='o')
plt.show()

# Added circular markers to each data point
```

# Add title and labels

```python
plt.plot(x, y, marker='o')

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Added title and axis labels to the graph
```

# Styling Plots

# Dashed red line

```python
plt.plot(x, y, linestyle='--', color='red')
plt.show()

# Changed line style to dashed and color to red
```

# Green square markers

```python
plt.plot(x, y, marker='s', markersize=10, color='green')
plt.show()

# Used square markers with larger size and green color
```

# Extra marker styles (experiment)

```python
plt.plot(x, y, marker='^', color='purple')
plt.show()

# Used triangle marker style with purple color
```

```python
plt.plot(x, y, marker='D', color='orange')
plt.show()


# Used diamond marker style with orange color
```
