# Exercise 2.C Working with multiple DataFrames

# Import Libraries

```python
import numpy as np
import pandas as pd

# Imported NumPy and pandas for data analysis
```

# Basic Aggregations
# Create df_sales

```python
df_sales = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'],
    'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
})

df_sales

# Created sales dataset with product, region, units sold, and price
```

# Add Revenue column

```python
df_sales["Revenue"] = df_sales["Units_Sold"] * df_sales["Price"]
df_sales


# Calculated Revenue by multiplying Units_Sold and Price
```

# Total values

```python
df_sales["Units_Sold"].sum(), df_sales["Revenue"].sum()

# Calculated total units sold and total revenue
```

# Mean values

```python
df_sales["Units_Sold"].mean(), df_sales["Revenue"].mean()

# Calculated average units sold and average revenue
```

# Grouping Data
# Group by Product

```python
df_sales.groupby("Product")[["Units_Sold", "Revenue"]].sum()

# Grouped by product and calculated total sales and revenue
```

# Group by Region

```python
df_sales.groupby("Region")[["Units_Sold", "Revenue"]].mean()

# Grouped by region and calculated average values
```

# Group by Product + Region

```python
df_sales.groupby(["Product", "Region"])[["Units_Sold", "Revenue"]].sum()

# Grouped by product and region together and calculated totals
```

# Sorting and Ranking Data
# Sort by Units_Sold (ascending)

```python
df_sales.sort_values("Units_Sold", ascending=True)

# Sorted data by lowest to highest units sold
```

# Sort by Revenue (descending)

```python
df_sales.sort_values("Revenue", ascending=False)

# Sorted data by highest revenue first
```

# Add Revenue Rank

```python
df_sales["Revenue_Rank"] = df_sales["Revenue"].rank(ascending=False)
df_sales

# Ranked revenue values (1 = highest revenue)
```

# Merging DataFrames
# Create df_products

```python
df_products = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Category': ['Electronics', 'Furniture', 'Appliances']
})
df_products

# Created product category lookup table
```

# Merge DataFrames

```python
df_merged = pd.merge(df_sales, df_products, on="Product")
df_merged

# Merged sales data with product categories
```

# Group by Category

```python
df_merged.groupby("Category")["Revenue"].sum()

# Calculated total revenue by product category
```
