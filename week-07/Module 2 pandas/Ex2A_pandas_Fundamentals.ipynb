# Series
# Create Python objects

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}

ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))
ser4 = pd.Series(ad)

print(ser1)
print(ser2)
print(ser3)
print(ser4)

# TypeError: Series() takes at most 2 positional arguments (4 given)
# This happens because multiple values must be passed as a list or array
```

# TypeError example

```python
ser5 = pd.Series(17, 18, 19, 20)


# TypeError:
# Series.__init__() takes from 1 to 4 positional arguments but 5 were given
```
