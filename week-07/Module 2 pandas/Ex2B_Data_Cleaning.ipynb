# Import Libraries

```python
import numpy as np
import pandas as pd

# Imported NumPy and pandas for data cleaning tasks
```

# Load Dataset

```python
df = pd.read_csv("test_data.csv")
df

# Loaded test_data.csv into a DataFrame
```

# Data Summary

```python
df.describe()

# Shows statistical summary of numeric columns (Age, Score, etc.)
```

# Filtering a DataFrame
# Age greater than 18

```python
df[df["Age"] > 18]

# Filters rows where Age is greater than 18
```

# Score between 50 and 100

```python
df[(df["Score"] > 50) & (df["Score"] < 100)]

# Filters rows where Score is between 50 and 100
```

# Modifying a DataFrame
# Remove invalid scores

```python
df_cleaned = df[df["Score"] >= 0]
df_cleaned


# Removed rows where Score is less than 0 (invalid data)
```

# Fill missing Age values

```python
df_cleaned["Age"] = df_cleaned["Age"].fillna(df_cleaned["Age"].mean())
df_cleaned

# Replaced missing Age values with mean Age
```

# Replace Score value of 0 with mean

```python
df_cleaned["Score"] = df_cleaned["Score"].replace(
    0,
    df_cleaned["Score"].mean()
)
df_cleaned

# Replaced Score value of 0 with average Score
```

# Drop remaining NaN values

```python
df_cleaned = df_cleaned.dropna()
df_cleaned

# Removed any remaining missing values
```

# Set ID as index and drop column

```python
df_cleaned = df_cleaned.set_index("ID")
df_cleaned

# Set ID as index and removed ID column from DataFrame
```
