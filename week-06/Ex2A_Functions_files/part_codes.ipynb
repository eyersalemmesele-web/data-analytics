```python
def get_supplier_code(code):
    supplier = code.split(":")[0]
    return supplier
# i used split to separate the string at ":" and return the supplier code
```

```python
print(get_supplier_code("ACME:123-L"))
```

```python
def get_product_num(code):
    start = code.find(":")
    end = code.find("-")
    product = code[start+1:end]
    return product
# i used find and slicing to get the product number between ":" and "-"
```

```python
print(get_product_num("ACME:123-L"))
```

```python
def get_size(code):
    size = code.split("-")[1]
    return size
# i used split with "-" and returned the size value
```

```python
print(get_size("ACME:123-L"))
```

```python
acme = "ACME:123-L"
di = "DI:12345-M"
ace = "ACE:1-12"
# i created variables for the different part codes
```

```python
supplier = get_supplier_code(acme)
part_num = get_product_num(acme)
size = get_size(acme)

print(f"For {acme}, the supplier is {supplier}. Part number {part_num} is available in size {size}.")
```

```python
supplier = get_supplier_code(di)
part_num = get_product_num(di)
size = get_size(di)

print(f"For {di}, the supplier is {supplier}. Part number {part_num} is available in size {size}.")
```

```python
supplier = get_supplier_code(ace)
part_num = get_product_num(ace)
size = get_size(ace)

print(f"For {ace}, the supplier is {supplier}. Part number {part_num} is available in size {size}.")
```

```python
#
codes = [acme, di, ace]

for code in codes:
    supplier = get_supplier_code(code)
    part_num = get_product_num(code)
    size = get_size(code)

    print(f"For {code}, the supplier is {supplier}. Part number {part_num} is available in size {size}.")
    
# i used a loop with a list so the program can work with multiple part codes automatically
```
