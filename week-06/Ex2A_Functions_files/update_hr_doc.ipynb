```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
print(hr_list)

# i created a list of tuples that stores employee id, department, name, and salary
# i updated Mark's last name from Blick to Blick-Hawley
# printing the list to make sure Mark's name was updated correctly
```

```python
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
print(hr_list)

# i updated Jim's department from OPS to CS and changed the salary to 60000
# printing the list again to confirm both updates stayed in the list
```

```python
for emp in hr_list:
    emp_id = emp[0]
    dept = emp[1]
    name = emp[2]
    salary = int(emp[3])

    print(f"{emp_id} | {dept} | {name} | {salary:,}")

# i used a loop and formatted the salary with commas to make the output easier to read
```
