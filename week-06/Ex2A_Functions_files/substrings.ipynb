```python
test_name = "Eyerusalem Debero"
# i made a variable to store my full name for testing
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == " ":
            first = name[:i]
            last = name[i+1:]

            print(f"Full name: {name}")
            print(f"First name: {first}")
            print(f"Last name: {last}")

# i created a function that loops through the name and split it into first and last name using space
```

```python
parse_and_display(test_name)
# i called the function using my variable to see the output.
```

```python
test_list = ["Mina Saeday", "Nobel haylu", "feven Semere"]
# i made a list of names to test my function with multiple values
```

```python
parse_and_display(test_list[0])

# testing first name from the list
```

```python
parse_and_display(test_list[1])

# testing second name from the list
```

```python
parse_and_display(test_list[2])
```

```python
def parse_and_display2(name):
    parts = name.split()

    if len(parts) == 1:
        print(f"Full name: {name}")
        print(f"Name: {parts[0]}")

    elif len(parts) == 2:
        print(f"Full name: {name}")
        print(f"First name: {parts[0]}")
        print(f"Last name: {parts[1]}")

    else:
        print(f"Full name: {name}")
        print(f"First name: {parts[0]}")
        print(f"Middle name: {' '.join(parts[1:-1])}")
        print(f"Last name: {parts[-1]}")

# i made a better function that can handle one word, two words, and three words names
```

```python
parse_and_display2("Eyerusalem Debero")
parse_and_display2("Mina Saeday")
parse_and_display2("Nobel haylu")

# i made a better function that can handle one word, two words, and three words names
```
