```python
import numpy as np
```

```python
# Array Basics
```

```python
fruits = ["apple", "banana", "cherry", "date"]
arr_1 = np.array(fruits)

print(arr_1)

# i created a one-dimensional numpy array from the fruits list
# this is a regular python list containing fruit names
```

```python
print(arr_1.dtype)

# the dtype returned is a string data type
# the letter shows the type of data stored
# the number shows the maximum length of the string
```

```python
arr_2 = np.array([
    [10, 11, 12],
    [13, 14, 15]
])

print(arr_2)

# i created a two-dimensional array
```

```python
print(arr_2.T)

# this prints the transpose of arr_2
# using .T does not permanently change arr_2
```

```python
print(arr_2)

# arr_2 is still the same after using transpose
```

```python
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

print(arr_3)

# i created another two-dimensional array
```

```python
arr_4 = np.array([
    arr_3[0, 2],
    arr_3[1, 1]
])

print(arr_4)

# this array uses slicing/index positions from arr_3
```

```python
arr_5 = np.ones((2, 2, 2))

print(arr_5)

# i created a three-dimensional array filled with ones
# each dimension contains two values
```
