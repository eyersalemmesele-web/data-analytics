```python
sales_data = [
('Laptop', 'East', 1200.00),
('Phone', 'West', 850.00),
('Tablet', 'East', 400.00),
('Monitor', 'South', 650.00),
('Keyboard', 'West', 75.00),
('Laptop', 'South', 1150.00),
('Phone', 'East', 900.00),
('Tablet', 'West', 425.00),
('Monitor', 'East', 680.00),
('Keyboard', 'South', 80.00),
('Laptop', 'West', 1300.00),
('Phone', 'West', 820.00),
]

# i created a sales data list with tuples that contain the product, region, and sale amount
```

```python
# Lab 1. Total sales by region
```

```python
def region_total(data, region):
    return sum(amt for _, r, amt in data if r == region)

print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))


# i used a generator expression to add all sale amounts for a specific region
# i tested the function with all three regions
```

```python
# Lab 2. Count of sales by region
```

```python
def region_count(data, region):
    return sum(1 for _, r, amt in data if r == region)

print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))

# i counted how many records belong to each region
# i tested the region count function
```

```python
# Lab 3. Average sale amount by region
```

```python
def region_average(data, region):

    total = region_total(data, region)
    count = region_count(data, region)

    if count == 0:
        return 0

    return total / count

print(region_average(sales_data, 'East'))
print(region_average(sales_data, 'West'))
print(region_average(sales_data, 'South'))
print(region_average(sales_data, 'North'))

# i used the total and count functions to calculate the average sale amount
# i also tested a region that does not exist to return 0
```

```python
# Lab 4. Filter by sale amount
```

```python
def sales_above(data, threshold):

    return [record for record in data if record[2] > threshold]

results = sales_above(sales_data, 500)

for record in results:
    print(record)

# i used list comprehension to return only sales above the threshold
# i tested the function using a threshold of 500
```

```python
def sales_below(data, threshold):

    return [record for record in data if record[2] < threshold]

results = sales_below(sales_data, 500)

for record in results:
    print(record)

# i created another function to return sales below the threshold
# i tested the function with sales below 500
```

```python
# Lab 5. Top and bottom sale
```

```python
def top_sale(data):

    return max(data, key=lambda s: s[2])

# i used max with lambda to find the highest sale amount


def bottom_sale(data):

    return min(data, key=lambda s: s[2])

print(top_sale(sales_data))
print(bottom_sale(sales_data))



# i used min with lambda to find the lowest sale amount
# i tested both functions
```

```python
# Lab 6. Apply a discount
```

```python
def apply_discount(data, discount_pct):

    return list(map(lambda s: (s[0], s[1], s[2] * (1 - discount_pct)), data))

discounted = apply_discount(sales_data, 0.15)

print("Before Discount:")
print(sales_data[:3])

print("After Discount:")
print(discounted[:3])

# i used map and lambda to reduce all sale amounts by the discount percentage
# i tested the function using a 15 percent discount
```

```python
# Lab 7. Unique regions
```

```python
def unique_regions(data):

    return sorted(set(region for _, region, _ in data))

print(unique_regions(sales_data))


# i used set to remove duplicates and sorted to organize the region names
# i tested the function to display all unique regions
```

```python
# Lab 8. Region summary
```

```python
def region_summary(data, region):

    total = region_total(data, region)
    count = region_count(data, region)
    average = region_average(data, region)

    region_data = [record for record in data if record[1] == region]

    top = top_sale(region_data)

    print(f"=== {region} Region ===")
    print(f"Records: {count}")
    print(f"Total Sales: ${total:,.2f}")
    print(f"Average Sale: ${average:,.2f}")
    print(f"Top Sale: {top[0]} — ${top[2]:,.2f}")

region_summary(sales_data, 'East')
region_summary(sales_data, 'West')
region_summary(sales_data, 'South')



# i combined multiple functions together to create a region summary report
# i tested the summary function with all regions
```

```python
# Lab 9. Full summary
```

```python
for region in unique_regions(sales_data):

    region_summary(sales_data, region)

# i used a loop with the unique regions function so the summaries print automatically
```
