```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062

# i created a function to calculate social security tax (6.2%)
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145

# i created a function to calculate medicare tax (1.45%)
```

```python
def get_federal_tax(gross_pay, code):

    if code == 0:
        rate = 0.23
    elif code == 1:
        rate = 0.21
    elif code == 2:
        rate = 0.195
    elif code == 3:
        rate = 0.185
    else:
        extra = code - 4
        rate = 0.18 - (0.005 * extra)

    return gross_pay * rate

# i created a function that calculates federal tax using withholding codes
```

```python
gross_pay = 750
code = 0

print("Social Security:", get_soc_sec_tax(gross_pay))
print("Medicare:", get_medicare_tax(gross_pay))
print("Federal Tax:", get_federal_tax(gross_pay, code))

# i tested the functions with person 1
```
