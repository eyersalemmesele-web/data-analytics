```python
def favorite_things(name, movie, music):
    return f"My name is {name}\nMy favorite movie is {movie}\nMy favorite music is {music}"
print(favorite_things("Eyerusalem Debero", "Black Panther", "Goodness of God"))

#i called the function using my name, favorite movie, and favorite song
#i created a function that returns my favorite things
```

```python
def why_im_here(purpose):
    return purpose
print(why_im_here("I joined Year Up to learn data analytics and build my career in tech."))

#i printed my purpose for joining the program
#i created a function to explain why i joined Year Up
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}"
print(favorite_place("Dallas", "I like the opportunities and lifestyle"))

#i tested my function with Dallas
#i created a function about my favorite place
```

```python
print(favorite_place("Eritrea", "It is my home and very meaningful to me"))
#i tested my function with Eritrea because it is my home and very meaningful to me
```
