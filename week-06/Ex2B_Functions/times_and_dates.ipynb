```python
import datetime
```

```python
today = datetime.datetime.now()

# i created a variable to store current date and time
```

```python
print("Today is:", today.strftime("%A, %B %d, %Y"))

# i used strftime to format today into readable format
```

```python
birthday = datetime.date(2001, 7, 21)
print("My birthday is:", birthday.strftime("%m/%d/%Y"))

# i created a date object using my real birthday and formatted it
```

```python
ninety_d = today + datetime.timedelta(days=90)
print("90 days from today is:", ninety_d.strftime("%B %d, %Y"))

# i used timedelta to add 90 days to today's date
```

```python
dinner_time = datetime.time(19, 30)
print("Let’s meet for dinner at:", dinner_time.strftime("%I:%M %p"))

# i used time and strftime to display time in AM/PM format
```
